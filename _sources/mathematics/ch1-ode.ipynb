# Решение дифференциальных уравнений и их систем

Для начала рассмотрим решение обыкновенного дифференциального уравнения (ОДУ), имеющего известное аналитическое решение.
Затем проинтегрируем ДУ с ненулевой правой частью.
Далее покажем способ численного решения системы дифференциальных уравнений (ДУ).

## Система ОДУ

Рассмотрим численное интегрирование системы ОДУ.
Система ОДУ получается всякий раз, когда необходимо решить ОДУ второго и более порядка.

Рассмотрим ОДУ, которое описывае динамику горизонтальных колебаний бруска под действием пружины жёсткостью $k$ при наличии трения его о горизонтальную поверхность:

$$
\ddot{x} + 2 {\rm\beta} \dot{x} + {\rm\omega}_0^2 x = 0,
$$(ode_eq)

где
$x$ - отклонение бруска от положения равновесия $x_0$ (когда пружина не деформирована);
${\rm\beta}$ - показатель затухания:

$$
{\rm\beta} = \cfrac{r}{2 m};
$$(beta_eq)

$r$ - коэффициент сопротивления среды;
$m$ - масса бруска;
${\rm\omega_0}$ - собственная круговая частота колебаний системы:

$$
{\rm\omega}_0^2 = k / m.
$$(omega_eq)

Решением этого уравнения является функция координаты от времени вида

$$
x(t) = A_0 e^{-{\rm\beta} t} \cos{({\rm\omega}_0 t + {\rm\varphi})},
$$(ode_sol_eq)

где
$A_0$ - амплитуда собственных колебаний;
$\rm \varphi$ - начальная фаза колебаний.
