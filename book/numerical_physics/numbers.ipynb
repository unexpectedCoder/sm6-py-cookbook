## Ошибки округления

Любые целочисленные арифметические операции абсолютно точны.
Единственное, что может случиться с целыми числами - это _переполнение_, выход результата операции за пределы диапазона.
С вещественными числами всё гораздо сложнее.
Помимо переполнения, им грозит потеря точности вблизи нуля, ошибки округления при арифметических операциях и другие ошибки, например, .

Далее в главе рассмотрим только примеры, показывающие особенности работы с вещественными числами на компьютере.
Всё сказанное не зависит от языка программирования - арифметические операции зависят от архитектуры процессора, а конкретнее от устройства арифметико-логического устройства.
